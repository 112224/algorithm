```python
#https://www.acmicpc.net/problem/1450
#냅색문제

from collections import deque
import sys
input = sys.stdin.readline

import bisect

n, c = map(int, input().split())
wei = list(map(int, input().split()))

we2 = wei[:n//2]
wei = wei[n//2:]

gr1, gr2 = [], []
def get_sums(a, gr, idx, val):
    if idx == len(a):
        gr.append(val)
        return

    get_sums(a,gr,idx+1,val)
    get_sums(a,gr,idx+1, val + a[idx])

get_sums(wei,gr1,0,0)
get_sums(we2,gr2,0,0)

gr1.sort()
gr2.sort()

ans = 0
for ele in gr1:
    a = bisect.bisect_right(gr2,c - ele)
    ans += a

print(ans)

##https://docs.python.org/ko/3/library/bisect.html
##bisect 관련 숙지하고 있을 것.
```

```python
#https://www.acmicpc.net/problem/16234
#인구 이동

from collections import deque
import sys
input = sys.stdin.readline

n, l, r = map(int, input().split())

ary = [list(map(int, input().split())) for _ in range(n)]

dx, dy = [0,0,1,-1], [1,-1,0,0]


def get_union(ary, l, r):

    ans = 0
    moved, once = False, True

    while moved or once:
        once = False
        moved = False
        up, up_val = [], []
        visit = [[False] * n for _ in range(n)]
        for i in range(n):
            for j in range(n):
                if visit[i][j]:
                    continue
                q = deque()
                q.append((i,j))
                uni = ary[i][j]
                uni_pos = [(i,j)]
                visit[i][j] = True
                while q:
                    x, y= q.popleft()
                    for k in range(4):
                        nx, ny = x+dx[k], y+dy[k]
                        if nx>=n or ny>=n or nx<0 or ny<0:
                            continue
                        if r >= abs(ary[x][y] - ary[nx][ny]) >= l and not visit[nx][ny]:
                            moved = True
                            visit[nx][ny] = True
                            uni_pos.append((nx,ny))
                            q.append((nx,ny))
                            uni += ary[nx][ny]
                if uni>ary[i][j]:
                    up.append(uni_pos)
                    up_val.append(uni)
        if moved:
            ans+=1
        for i in range(len(up)):
            val = up_val[i]//len(up[i])
            for k in up[i]:
                x,y = k
                ary[x][y] = val
    return ans


print(get_union(ary,l,r))
```

```python
#https://www.acmicpc.net/problem/3665
#최종 순위

#include <iostream>
#include <algorithm>
#include <vector>
#include <queue>
#include <string>
#include <cstring>
using namespace std;


int main()
{
	ios::sync_with_stdio(false);
	cin.tie(0), cout.tie(0);
	
	int t; cin >> t;

	while (t--)
	{
		int n; cin >> n;
		vector<vector<int> >ar(n + 1);
		vector<int>ind(n + 1, 0);
		int prev = 0; cin >> prev;

		for (int i = 1; i < n; i++) 
		{
			int tmp; cin >> tmp;
			ar[tmp] = ar[prev];
			ar[tmp].push_back(prev);
			ind[prev]=n-i;
			prev = tmp;
		}
		int m; cin >> m;
		while (m--)
		{
			int a, b; cin >> a >> b;
			bool flag = false;
			for (auto it = ar[a].begin(); it != ar[a].end(); ++it)
			{
				if (*it == b)
				{
					flag = true;
					ar[a].erase(it);
					ind[b]--;
					ar[b].push_back(a);
					ind[a]++;
					break;
				}
			}
			if (!flag)
			{
				for (auto it = ar[b].begin(); it != ar[b].end(); ++it)
				{
					if (*it == a)
					{
						ar[b].erase(it);
						ind[a]--;
						ar[a].push_back(b);
						ind[b]++;
						break;
					}
				}
			}
		}

		queue<int>q;
		vector<int>ans;
		for (int i = 1; i <= n; i++)
			if (ind[i] == 0)
				q.push(i);

		bool flag = false;
		while (!q.empty())
		{
			if (q.size() == 2)
			{
				cout << "IMPOSSIBLE\n";
				flag = true;
				break;
			}
			int now = q.front(); q.pop();
			ans.push_back(now);

			for (int k : ar[now])
			{
				if ((--ind[k]) == 0)
					q.push(k);
			}
		}
		
		if (!flag)
		{
			if (ans.size() == n)
			{
				for (int i = n - 1; i >= 0; i--)
					cout << ans[i] << ' ';
				cout << '\n';
			}
			else
			{
				cout << "IMPOSSIBLE\n";
				flag = true;
			}

		}
	}
	
}
```

```python
#https://www.acmicpc.net/problem/20040
#사이클 게임

from collections import deque
import sys
input = sys.stdin.readline

n,m = map(int, input().split())
pare = [i for i in range(n)]


def find(x):
    if pare[x] == x:
        return x
    pare[x] = find(pare[x])
    return pare[x]

def uni(x, y):
    a, b = find(x), find(y)
    if a != b:
        pare[a] = b

for i in range(1, m+1):
    a, b = map(int, input().split())
    if find(a) == find(b):
        print(i)
        break
    else:
        uni(a,b)
    if i == m:
        print(0)
```

```python
#https://www.acmicpc.net/problem/20040
#사이클 게임

from collections import deque
import bisect
import sys
input = sys.stdin.readline

n = int(input())
ary = list(map(int, input().split()))
ary.sort()
x = int(input())

ans = 0
for i in range(n):
    a = bisect.bisect_left(ary,x - ary[i],lo=i)
    b = bisect.bisect_right(ary, x-ary[i],lo = i)
    for j in range(a,b):
        if x == ary[i] + ary[j]:
            ans += 1

print(ans)
```
