```python
#https://www.acmicpc.net/problem/14002
#가장 긴 증가하는 부분 수열 4

from collections import deque
import bisect
import sys
input = sys.stdin.readline

n = int(input())

ary = list(map(int, input().split()))
lis = [ary[0]]
idx = [0]*n

for i in range(n):
    if lis[-1] < ary[i]:
        lis.append(ary[i])
        idx[i] = len(lis) - 1
    else:
        j = bisect.bisect_left(lis,ary[i])
        idx[i] = j
        lis[j] = ary[i]

ans = len(lis)
print(ans)

st = []
for i in range(n-1,-1,-1):
    if ans - 1 == idx[i]:
        st.append(ary[i])
        ans -= 1
while st:
    print(st.pop(), end=' ')
```

```python
#https://www.acmicpc.net/problem/3980
#선발 명단

from collections import deque
import bisect
import sys
input = sys.stdin.readline

tc = int(input())
full = (1<<11) - 1


def solve(filled, idx, ary, dp, ret):

    if filled == full:
        return ret
    if dp[idx][filled] != 0:
        return dp[idx][filled]

    for i in range(11):
        if filled & (1<<i) == 0 and ary[idx][i] != 0:
            n_filled = filled | (1<<i)
            dp[idx][n_filled] = max(dp[idx][n_filled],solve(n_filled, idx + 1, ary, dp, ret+ary[idx][i]))

    return max(dp[idx])


while tc > 0:
    tc -= 1
    ary = [list(map(int, input().split())) for _ in range(11)]
    dp = [[0]*(1<<12) for _ in range(12)]
    print(solve(0,0,ary,dp,0))
```

```python
#https://www.acmicpc.net/problem/1577
#도로의 개수

from collections import deque
import bisect
import sys
input = sys.stdin.readline

n, m = map(int, input().split())
k = int(input())

don = [[[True]*2 for _ in range(m+1)] for i in range(n+1)]

for i in range(k):
    a,b,c,d = map(int, input().split())
    if a==c:
        don[a][max(b,d)][0] = False
    else:
        don[max(a,c)][b][1] = False

dp = [[-1]*(m+1) for _ in range(n+1)]


def solve(x, y):
    if x==0 and y==0:
        return 1
    if x<0 or y<0:
        return 0
    if dp[x][y] != -1:
        return dp[x][y]
    dp[x][y] = 0
    if don[x][y][1]:
        dp[x][y] += solve(x-1,y)
    if don[x][y][0]:
        dp[x][y] += solve(x,y-1)
    return dp[x][y]


print(solve(n,m))
```

```python
#https://www.acmicpc.net/problem/2629
#양팔저울

from collections import deque
import bisect
import sys
input = sys.stdin.readline

n = int(input())
ary = list(map(int, input().split()))
k = int(input())
cn = list(map(int,input().split()))

dp = [[-1]*40010 for _ in range(40)]
weight = [False]*40010


def solve(idx, val, n):
    if dp[idx][val] != -1:
        return
    dp[idx][val] = 1
    weight[val] = True
    if idx == n:
        return
    solve(idx + 1, val + ary[idx], n)
    solve(idx + 1, val, n)
    solve(idx + 1, abs(val - ary[idx]), n)



solve(0,0,n)
ans = []
for c in cn:
    if weight[c]:
        ans.append("Y")
    else:
        ans.append("N")
print(' '.join(ans))
```
