```python
#https://www.acmicpc.net/problem/1339
#단어 수학

import sys

input = sys.stdin.readline

n = int(input())
visit = [0]*26
vals = [0]*26
a = [list(input().strip()) for _ in range(n)]
for now in a:
    tl = len(now)
    idx = 0
    for a in now:
        i = ord(a) - ord('A')
        vals[i] += (10 ** (tl - 1 - idx))
        idx += 1
vals.sort(reverse=True)
ans = 0
i = 9
for val in vals:
    if val == 0:
        break
    ans += val * i
    i -= 1
print(ans)
```

```python
#https://www.acmicpc.net/problem/14888
#연산자 끼워넣기

import sys

input = sys.stdin.readline

n = int(input())
val = list(map(int, input().split()))
cnts = list(map(int, input().split()))

global m,M
m = 1000000000
M = -1000000000
def cal(cnts, val, idx, ret):
    if sum(cnts) == 0:
        global m,M
        m = min(m, ret)
        M = max(M, ret)
        return

    for i in range(4):
        if cnts[i]:
            cnts[i] -= 1
            if i == 0:
                cal(cnts, val, idx+1, ret + val[idx+1])
            elif i == 1:
                cal(cnts,val,idx+1,ret - val[idx+1])
            elif i == 2:
                cal(cnts,val,idx+1, ret * val[idx+1])
            elif i == 3:
                if ret<0:
                    cal(cnts, val, idx + 1, -(-ret // val[idx + 1]))
                else:
                    cal(cnts, val, idx + 1, ret // val[idx + 1])
            cnts[i] += 1


cal(cnts, val, 0, val[0])
print(M)
print(m)
```

```python
#https://www.acmicpc.net/problem/14889
#스타트와 링크

from itertools import combinations
def cal(senergy, a, b):
    n = len(a)
    ret_a = 0
    ret_b = 0
    for i in range(n-1):
        for j in range(i+1,n):
            ret_a += senergy[a[i]][a[j]]
            ret_a += senergy[a[j]][a[i]]
            ret_b += senergy[b[i]][b[j]]
            ret_b += senergy[b[j]][b[i]]

    return abs(ret_b - ret_a)


def check(ary):
    n = len(ary)
    tmp = [i for i in range(n)]
    s1 = set(tmp)
    s2 = combinations(tmp, n//2)

    ans = 987654321
    for ele in s2:
        s3 = s1 - set(ele)
        tmp = cal(ary,list(ele),list(s3))
        ans = min(ans, tmp)
    return ans


import sys
input = sys.stdin.readline

n = int(input())
a = [list(map(int, input().split())) for _ in range(n)]

print(check(a))
```

```python
#https://www.acmicpc.net/problem/6603
#로또

from itertools import combinations
import sys
input = sys.stdin.readline


while True:
    a = list(map(int, input().split()))
    if a[0] == 0:
        break
    b = a[1:]
    ans = combinations(b,6)
    for ret in ans:
        print(' '.join(map(str, ret)))
    print()
```

```python
#https://www.acmicpc.net/problem/1182
#부분수열의 합

from itertools import combinations
import sys
input = sys.stdin.readline

n, m = map(int, input().split())
a = list(map(int, input().split()))
ans = 0
for i in range(1, n+1):
    combi = combinations(a, i)
    for ele in combi:
        if sum(ele) == m:
            ans += 1
print(ans)
```

```python
#https://www.acmicpc.net/problem/14225
#부분수열의 합

import sys
input = sys.stdin.readline

n = int(input())
a = list(map(int, input().split()))
visit = [False] * (100000*20 + 10)


def cal(idx, sum):
    if idx == n:
        visit[sum] = True
        return
    cal(idx+1, sum + a[idx])
    cal(idx+1, sum)


cal(0,0)
for i in range(len(visit)):
    if not visit[i]:
        print(i)
        break
```

```python
#https://www.acmicpc.net/problem/15658
#연산자 끼워넣기 (2)

import sys

input = sys.stdin.readline

n = int(input())
val = list(map(int, input().split()))
cnts = list(map(int, input().split()))

global m,M
m = 1000000000
M = -1000000000
def cal(cnts, val, idx, ret):
    if idx == n-1:
        global m,M
        m = min(m, ret)
        M = max(M, ret)
        return

    for i in range(4):
        if cnts[i]:
            cnts[i] -= 1
            if i == 0:
                cal(cnts, val, idx+1, ret + val[idx+1])
            elif i == 1:
                cal(cnts,val,idx+1,ret - val[idx+1])
            elif i == 2:
                cal(cnts,val,idx+1, ret * val[idx+1])
            elif i == 3:
                if ret<0:
                    cal(cnts, val, idx + 1, -(-ret // val[idx + 1]))
                else:
                    cal(cnts, val, idx + 1, ret // val[idx + 1])
            cnts[i] += 1
cal(cnts, val, 0, val[0])
print(M)
print(m)
```

```python
#https://www.acmicpc.net/problem/14500
#테트로미노

#https://www.acmicpc.net/problem/14500
#테트로미노

import sys
input = sys.stdin.readline

n, m = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]
visit = [[False]*m for _ in range(n)]

move = [(1,0),(-1,0),(0,1),(0,-1)]
global ans
ans = -1


def go(x, y, cnt, ret):
    if cnt == 3:
        global ans
        ans = max(ans, ret)
        return
    for k in move:
        nx = x + k[0]
        ny = y + k[1]
        if nx<0 or ny<0 or nx>=n or ny>=m:
            continue
        if not visit[nx][ny]:
            visit[nx][ny] = True
            go(nx, ny, cnt +1, ret + a[nx][ny])
            visit[nx][ny] = False


# + 모양에서 한 귀퉁이 떼내기
def go1(x,y):
    global ans
    ret = a[x][y]
    cnt = 0
    for k in move:
        nx = x + k[0]
        ny = y + k[1]
        if nx < 0 or ny < 0 or nx >= n or ny >= m:
            continue
        ret += a[nx][ny]
        cnt += 1
    if cnt == 3:
        ans = max(ans, ret)
    elif cnt > 3:
        for k in move:
            nx = x + k[0]
            ny = y + k[1]
            ans = max(ans, ret - a[nx][ny])


for i in range(n):
    for j in range(m):
        visit[i][j] = True
        go(i,j,0,a[i][j])
        go1(i,j)
        visit[i][j] = False
        
print(ans)
```
