```python
#include<iostream>
#include<queue>
#include<algorithm>
#include<vector>
using namespace std;

typedef struct ii
{
	int b, d;
}ii;

bool operator <(ii u, ii v) {
	return u.d < v.d;
}
int dist[1001];
int pre[1001];

int main()
{
	ios::sync_with_stdio(false);
	cin.tie(0); cout.tie(0);

	int n, m; cin >> n >> m;
	vector<vector<ii> >adj(n + 1);

	for (int i = 2; i <= n; i++) {
		dist[i] = 987654321;
		pre[i] = -1;
	}
	for (int i = 0; i < m; i++) {
		int a, b, c; cin >> a >> b >> c;
		adj[a].push_back({ b,c });
		adj[b].push_back({ a,c });
	}
	priority_queue<ii> pq;
	pq.push({ 1,0 });
	while (!pq.empty()) {
		ii now = pq.top(); pq.pop();

		for (auto k : adj[now.b]) {
			if (dist[k.b] > dist[now.b] + k.d) {
				pre[k.b] = now.b;
				dist[k.b] = dist[now.b] + k.d;
				pq.push(k);
			}
		}
	}
	cout << n - 1 << '\n';
	for (int i = 2; i <= n; i++)
		cout << pre[i] << ' ' << i << '\n';
}
```

```python
#https://www.acmicpc.net/problem/6209
#제자리 멀리뛰기

from collections import deque
from queue import PriorityQueue as pq
import sys
input = sys.stdin.readline

d, n, m = map(int, input().split())

ary = [0] + [int(input()) for _ in range(n)]
ary.sort()
diff = [ary[i+1]-ary[i] for i in range(n)]

hi, lo = d, 0
ans = 0
while hi>=lo:
    dif2 = diff.copy()
    mid = (hi+lo)//2
    cnt = 0
    for i in range(n):
        if dif2[i] >= mid:
            cnt+=1
        else:
            if i+1 != n:
                dif2[i+1] += dif2[i]
    if cnt == n-m:
        ans = mid
        lo = mid+1
    elif cnt < n-m:
        hi = mid-1
    else:
        lo = mid+1

print(ans)
```

```python
#https://www.acmicpc.net/problem/1516
#게임 개발

from collections import deque
from queue import PriorityQueue as pq
import sys
input = sys.stdin.readline

n = int(input())
ind = [0]*(n+1)
ins = [[] for _ in range(n+1)]
out = [[] for _ in range(n+1)]
build = [0]*(n+1)
for i in range(1,n+1):
    tmp = list(map(int,input().split()))
    build[i] += tmp[0]
    for ele in tmp[1:-1]:
        out[ele].append(i)
        ind[i] += 1
        ins[i].append(ele)

q = deque()

for i in range(1,n+1):
    if ind[i] == 0:
        q.append(i)

while q:
    now = q.popleft()

    for ne in out[now]:
        ind[ne] -= 1
        if ind[ne] == 0:
            q.append(ne)
            tmp = 0
            for vv in ins[ne]:
                tmp = max(tmp, build[vv])
            build[ne] += tmp

for i in range(1,n+1):
    print(build[i])
```
