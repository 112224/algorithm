```python
#https://www.acmicpc.net/problem/14620
#꽃길

import sys
input = sys.stdin.readline

n = int(input())
gar = [list(map(int, input().split())) for _ in range(n)]
dx = [0,1,-1,0,0]
dy = [0,0,0,1,-1]
visit = [[False]*n for _ in range(n)]


def solve(position, n):
    ret = 0
    ck = []
    for pos in position:
        x = pos//n
        y = pos%n

        if x==0 or x==n-1 or y ==0 or y==n-1:
            return 987654321
        for i in range(5):
            nx,ny = x+dx[i], y+dy[i]
            ck.append((nx,ny))
            ret += gar[nx][ny]
    if len(set(ck)) == 15:
        return ret
    else:
        return 987654321

ans = 987654321
for i in range(n*n):
    for j in range(n*n):
        for k in range(n*n):
            pos = [i,j,k]
            ans = min(ans, solve(pos, n))
print(ans)
```

```python
#https://www.acmicpc.net/problem/1707
#이분 그래프

import sys
input = sys.stdin.readline
from collections import deque

tc = int(input())

for _ in range(tc):
    v, e = map(int, input().split())
    adj = [[] for __ in range(v+1)]
    visit = [0]*v
    for __ in range(e):
        s, d = map(int, input().split())
        adj[s-1].append(d-1)
        adj[d-1].append(s-1)
    flag = True
    k = 0
    for k in range(v):
        if not flag:
            break
        if visit[k] == 0:
            q = deque()
            q.append(k)
            visit[k] = 1
            while q and flag:
                i = q.popleft()
                for x in adj[i]:
                    if visit[x] == visit[i]:
                        flag = False
                        break
                    elif visit[x] == 0:
                        visit[x] = -1*visit[i]
                        q.append(x)

    if flag:
        print("YES")
    else:
        print("NO")
```

```python
#https://www.acmicpc.net/problem/1806
#부분합

from collections import deque
import sys
input = sys.stdin.readline

n, s = map(int,input().split())

ary = list(map(int, input().split()))
b = [0]*(n+1)

for i in range(1,n+1):
    b[i] = b[i-1] + ary[i-1]

q = deque()
q.append((0,1))

ans = 987654321
while q:
    st,ed = q.popleft()
    if ed == st or ed>n:
        continue
    if b[ed] - b[st] >= s:
        ans = min(ans, ed-st)
        q.append((st+1,ed))
    else:
        q.append((st,ed+1))

print(ans)
```

```python
#https://www.acmicpc.net/problem/9376
#탈옥

from collections import deque

def bfs(x,y):
    dist = [[-1]*(w+2) for _ in range(h+2)]
    q = deque()
    q.append((x,y))
    dist[x][y]=0
    while q:
        x, y = q.popleft()
        for dx, dy in (-1,0),(1,0),(0,-1),(0,1):
            nx, ny = x+dx, y+dy
            if nx < 0 or nx > h+1 or ny < 0 or ny > w+1:
                continue
            if dist[nx][ny]>=0 or maze[nx][ny]=='*':
                continue
            if maze[nx][ny]=='#':
                dist[nx][ny]=dist[x][y]+1
                q.append((nx,ny))
            elif maze[nx][ny]=='.':
                dist[nx][ny]=dist[x][y]
                q.appendleft((nx,ny))

    return dist


for _ in range(int(input())):
    h, w = map(int, input().split())
    maze=[]
    maze.append(list('.'*(w+2)))
    for _ in range(h):
        maze.append(list('.'+input().strip()+'.'))
    maze.append(list('.'*(w+2)))
    dq = deque()
    for i in range(h+2):
        for j in range(w+2):
            if maze[i][j]=='$':
                maze[i][j]='.'
                dq.append((i,j))
    sx, sy = dq.popleft()
    d1 = bfs(sx, sy)
    sx, sy = dq.popleft()
    d2 = bfs(sx, sy)
    d3 = bfs(0, 0)
    ans = 1000000000
    for i in range(h+2):
        for j in range(w+2):
            if maze[i][j]=='*':
                continue
            k = d1[i][j] + d2[i][j] + d3[i][j]
            if maze[i][j]=='#':
                k-=2
            ans = min(k,ans)

    print(ans)
```

```python
#https://www.acmicpc.net/problem/7579
#앱

N, M = map(int, input().split())
A = [0] + list(map(int, input().split())) #byte
C = [0] + list(map(int, input().split())) #cost
dp = [[0 for _ in range(sum(C)+1)] for _ in range(N+1)] #냅색알고리즘이 실행될 dp
result = sum(C) #열의 최댓값

for i in range(1, N+1):
    byte = A[i]
    cost = C[i]
    
    for j in range(1, sum(C) + 1):
        if j < cost: #현재 앱을 비활성화할만큼의 cost가 충분하지 않을 경우
            dp[i][j] = dp[i-1][j]
        else:
        	#같은 cost 내에서 현재 앱을 끈 뒤의 byte와 현재 앱을 끄지 않은 뒤의 byte를 비교
            dp[i][j] = max(byte + dp[i-1][j-cost], dp[i-1][j])
            
        if dp[i][j] >= M: #조건이 충족된다면
            result = min(result, j) #더 작은 cost값으로 갱신
if M != 0:
    print(result)
else:
    print(0)
```
