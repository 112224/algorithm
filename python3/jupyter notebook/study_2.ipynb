```python
#https://www.acmicpc.net/problem/1987
#알파벳

import sys
input = sys.stdin.readline
#print = sys.stdout.write


r, c = map(int, input().split())
a = [list(input().strip()) for _ in range(r)]
visit = [False]*26

global ans
ans = -1
di = [(0,1),(1,0),(-1,0),(0,-1)]

def go(i,j,ret):
    flag = False
    for nn in di:
        ni = i + nn[0]
        nj = j + nn[1]
        if ni<0 or nj<0 or ni>=r or nj>=c:
            continue
        val = ord(a[ni][nj]) - ord('A')
        if not visit[val]:
            flag = True
            visit[val] = True
            go(ni,nj,ret+1)
            visit[val] = False
    if not flag:
        global ans
        ans = max(ans, ret)
        return

fi = ord(a[0][0]) - ord('A')
visit[fi] = True
go(0,0,1)

print(ans)
```

```python
#https://www.acmicpc.net/problem/14225
#부분 수열의 합

import sys
input = sys.stdin.readline
#print = sys.stdout.write


n = int(input())
a = list(map(int, input().split()))
c = [False]*(n*100000+10)
for i in range(1<<n):
    s = 0
    for j in range(n):
        if (i&(1<<j)):
            s += a[j]
    c[s] = True
i = 1
while True:
    if c[i] == False:
        break
    i += 1
print(i)
```

```python
#https://www.acmicpc.net/problem/1062
#가르침

import sys
input = sys.stdin.readline
#print = sys.stdout.write


n, k = map(int, input().split())
words = [0] * n
for i in range(n):
    tmp = input().strip()
    for c in tmp:
        words[i] |= (1 << (ord(c) - ord('a')))


def conut(mask):
    cnt = 0
    for word in words:
        if (word & ((1<<26)-1 - mask)) == 0:
            cnt += 1
    return cnt

def go(idx, rem, mask):
    if rem < 0:
        return 0

    if idx == 26:
        return conut(mask)
    ans = 0
    tmp = go(idx + 1, rem - 1, mask | (1 << idx))
    ans = max(ans, tmp)

    if idx not in [ord('a') - ord('a'), ord('n') - ord('a'), ord('t') - ord('a'),
               ord('i') - ord('a'), ord('c') - ord('a')]:
        tmp = go(idx + 1, rem, mask)
        ans = max(ans, tmp)
    return ans

print(go(0, k, 0))
```

```python
#https://www.acmicpc.net/problem/13460
#구슬 탈출 2

#비트마스크에 들어가 있는 이유
#10번 움직일 거 미리 정해서 넘겨줘 버리는 걸로 구현

import sys
input = sys.stdin.readline
#print = sys.stdout.write

dx = [0,0,1,-1]
dy = [1,-1,0,0]
LIMIT = 10
class Result:
    def __init__(self, moved, hole, x, y):
        self.moved = moved
        self.hole = hole
        self.x = x
        self.y = y

def gen(k):
    a = [0]*LIMIT
    for i in range(LIMIT):
        a[i] = (k&3)
        k >>= 2
    return a
def simulate(a, k, x, y):
    n = len(a)
    m = len(a[0])
    if a[x][y] == '.':
        return Result(False, False, x, y)
    moved = False
    while True:
        nx, ny = x+dx[k], y+dy[k]
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            return Result(moved, False, x, y)
        ch = a[nx][ny]
        if ch == '#':
            return Result(moved, False, x, y)
        elif ch in 'RB':
            return Result(moved, False, x, y)
        elif ch == '.':
            a[x][y],a[nx][ny] = a[nx][ny],a[x][y]
            x,y = nx,ny
            moved = True
        elif ch == 'O':
            a[x][y] = '.'
            moved = True
            return Result(moved, True, x, y)

def check(a, dirs):
    n = len(a)
    m = len(a[0])
    rx,ry = 0,0
    bx,by = 0,0

    for i in range(n):
        for j in range(m):
            if a[i][j] == 'R':
                rx,ry = i,j
            elif a[i][j] == 'B':
                bx,by = i,j

    cnt = 0
    for k in dirs:
        cnt += 1
        hole1 = hole2 = False
        while True:
            p1 = simulate(a, k, rx, ry)
            rx,ry = p1.x, p1.y
            p2 = simulate(a, k, bx, by)
            bx,by = p2.x, p2.y

            if not p1.moved and not p2.moved:
                break
            if p1.hole:
                hole1 = True
            if p2.hole:
                hole2 = True

        if hole2:
            return -1
        if hole1:
            return cnt

    return -1

def valid(dirs):
    l = len(dirs)
    for i in range(l-1):
        if dirs[i] == 0 and dirs[i+1] == 1:
            return False
        if dirs[i] == 1 and dirs[i+1] == 0:
            return False
        if dirs[i] == 2 and dirs[i+1] == 3:
            return False
        if dirs[i] == 3 and dirs[i+1] == 2:
            return False
        if dirs[i] == dirs[i+1]:
            return False
    return True

n,m = map(int,input().split())
original = [input().strip() for _ in range(n)]
ans = -1
for k in range(1<<(LIMIT*2)):
    dirs = gen(k)
    if not valid(dirs):
        continue
    a = [list(row) for row in original]
    cur = check(a, dirs)
    if cur == -1:
        continue
    if ans == -1 or ans > cur:
        ans = cur

print(ans)
```

```python
#https://www.acmicpc.net/problem/12100
#2048


import sys
input = sys.stdin.readline
#print = sys.stdout.write

dx = [0,0,1,-1]
dy = [1,-1,0,0]
LIMIT = 5
def gen(k):
    a = [0]*LIMIT
    for i in range(LIMIT):
        a[i] = (k&3)
        k >>= 2
    return a

def check(a, dirs):
    n = len(a)
    d = [row[:] for row in a]

    for dir in dirs:
        ok = False
        merged = [[False]*n for _ in range(n)]
        
        while True:
            ok = False
            if dir == 0:
                for i in range(n-2, -1, -1):
                    for j in range(n):
                        if d[i][j] == 0:
                            continue
                        if d[i+1][j] == 0:
                            d[i+1][j] = d[i][j]
                            merged[i+1][j] = merged[i][j]
                            d[i][j] = 0
                            ok = True
                        elif d[i+1][j] == d[i][j]:
                            if not merged[i][j] and not merged[i+1][j]:
                                d[i+1][j] *= 2
                                merged[i+1][j] = True
                                d[i][j] = 0
                                ok = True
            elif dir == 1:
                for i in range(1, n):
                    for j in range(n):
                        if d[i][j] == 0:
                            continue
                        if d[i-1][j] == 0:
                            d[i-1][j] = d[i][j]
                            merged[i-1][j] = merged[i][j]
                            d[i][j] = 0
                            ok = True
                        elif d[i-1][j] == d[i][j]:
                            if not merged[i][j] and not merged[i-1][j]:
                                d[i-1][j] *= 2
                                merged[i-1][j] = True
                                d[i][j] = 0
                                ok = True
            elif dir == 2:
                for j in range(1, n):
                    for i in range(n):
                        if d[i][j] == 0:
                            continue
                        if d[i][j-1] == 0:
                            d[i][j-1] = d[i][j]
                            merged[i][j-1] = merged[i][j]
                            d[i][j] = 0
                            ok = True
                        elif d[i][j-1] == d[i][j]:
                            if not merged[i][j] and not merged[i][j-1]:
                                d[i][j-1] *= 2
                                merged[i][j-1] = True
                                d[i][j] = 0
                                ok = True
            elif dir == 3:
                for j in range(n-2, -1, -1):
                    for i in range(n):
                        if d[i][j] == 0:
                            continue
                        if d[i][j+1] == 0:
                            d[i][j+1] = d[i][j]
                            merged[i][j+1] = merged[i][j]
                            d[i][j] = 0
                            ok = True
                        elif d[i][j+1] == d[i][j]:
                            if not merged[i][j] and not merged[i][j+1]:
                                d[i][j+1] *= 2
                                merged[i][j+1] = True
                                d[i][j] = 0
                                ok = True
            if not ok:
                break

    ans = max([max(row) for row in d])
    return ans

n = int(input())
a = [list(map(int,input().split())) for _ in range(n)]
ans = 0
for k in range(1<<(LIMIT*2)):
    dirs = gen(k)
    cur = check(a, dirs)
    if ans < cur:
        ans = cur
print(ans)
```

```python
#https://www.acmicpc.net/problem/16928
#뱀과 사다리 게임

import sys
input = sys.stdin.readline
#print = sys.stdout.write

from collections import deque

n, m = map(int, input().split())
board = list(range(101))
visit = [-1] * 101
for _ in range(n+m):
    st, ed = map(int, input().split())
    board[st] = ed

a = deque()
visit[1] = 0
a.append(1)


while a:
    x = a.popleft()
    for i in range(1,7):
        y = x + i
        if y > 100:
            continue
        y = board[y]
        if visit[y] == -1:
            visit[y] = visit[x] + 1
            a.append(y)
print(visit[100])
```

```python
#https://www.acmicpc.net/problem/16948
#데스 나이트

import sys
input = sys.stdin.readline
#print = sys.stdout.write

from collections import deque

n = int(input())

dr = [-2,-2,0,0,2,2]
dc = [-1,1,-2,2,-1,1]

r1, c1, r2, c2 = map(int, input().split())

visit = [[False]*n for _ in range(n)]

a = deque()
visit[r1][c1] = True
a.append((r1,c1,0))

ans = -1
while a:
    now = a.popleft()
    x = now[0]; y = now[1]; cnt = now[2]
    if x == r2 and y == c2:
        ans = cnt
        break

    for i in range(6):
        nr = x + dr[i]
        nc = y + dc[i]
        if nr<0 or nc<0 or nr>=n or nc >= n:
            continue
        if not visit[nr][nc]:
            visit[nr][nc] = True
            a.append((nr,nc, cnt + 1))

print(ans)
```

```python
#https://www.acmicpc.net/problem/12886
#돌 그룹

import sys
input = sys.stdin.readline
#print = sys.stdout.write

from collections import deque

a,b,c = map(int, input().split())
s = a + b + c

if s%3 != 0:
    print(0)
else:
    visit = [[False]*(3*s+1) for _ in range(3*s+1)]
    q = deque()
    q.append((a,b))
    flag = False
    while q:
        x, y = q.popleft()
        if x == y and s - 3*x == 0:
            flag = True
            break
        tmp = [x, y, s-x-y]
        for i in range(3):
            for j in range(3):
                if tmp[i] < tmp[j]:
                    tmp2 = [x, y, s-x-y]
                    tmp2[i] += tmp[i]
                    tmp2[j] -= tmp[i]
                    if not visit[tmp2[0]][tmp2[1]]:
                        visit[tmp2[0]][tmp2[1]] = True
                        q.append((tmp2[0],tmp2[1]))
    if flag:
        print(1)
    else:
        print(0)
```

```python
#https://www.acmicpc.net/problem/16946
#벽 부수고 이동하기 4

import sys
input = sys.stdin.readline
#print = sys.stdout.write

from collections import deque

dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]
n, m = map(int, input().split())
a = [list(map(int, list(input()))) for _ in range(n)]
group = [[-1] * m for _ in range(n)]
check = [[False] * m for _ in range(n)]
group_size = []


def bfs(sx, sy):
    g = len(group_size)
    q = deque()
    q.append((sx, sy))
    group[sx][sy] = g
    check[sx][sy] = True
    cnt = 1
    while q:
        x, y = q.popleft()
        for k in range(4):
            nx, ny = x + dx[k], y + dy[k]
            if 0 <= nx < n and 0 <= ny < m:
                if check[nx][ny] == False and a[nx][ny] == 0:
                    check[nx][ny] = True
                    group[nx][ny] = g
                    q.append((nx, ny))
                    cnt += 1

    group_size.append(cnt)


for i in range(n):
    for j in range(m):
        if a[i][j] == 0 and check[i][j] == False:
            bfs(i, j)

for i in range(n):
    for j in range(m):
        if a[i][j] == 0:
            print(0, end='')
        else:
            near = set()
            for k in range(4):
                nx, ny = i + dx[k], j + dy[k]
                if 0 <= nx < n and 0 <= ny < m:
                    if a[nx][ny] == 0:
                        near.add(group[nx][ny])
            ans = 1
            for g in near:
                ans += group_size[g]
            print(ans % 10, end='')
    print()
```

```python
#https://www.acmicpc.net/problem/14442
#벽 부수고 이동하기 2

import sys
input = sys.stdin.readline
#print = sys.stdout.write

from collections import deque

dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]

n,m,k = map(int, input().split())
board = [list(input().strip()) for _ in range(n)]
visit = [[[-1]*(k+1) for i in range(m+1)] for j in range(n+1)]

q = deque()
visit[0][0][k] = 1
q.append((0,0,k))

while q:
    x,y,z = q.popleft()
    for i in range(4):
        nx, ny = x+dx[i], y+dy[i]
        if nx<0 or ny<0 or nx>=n or ny>=m:
            continue
        if board[nx][ny] == '0' and visit[nx][ny][z] == -1:
            visit[nx][ny][z] = visit[x][y][z] + 1
            q.append((nx,ny,z))
        if z>0:
            if board[nx][ny] == '1' and visit[nx][ny][z] == -1:
                visit[nx][ny][z-1] = visit[x][y][z] + 1
                q.append((nx,ny,z-1))
ans = -1
for ele in visit[n-1][m-1]:
    if ele != -1:
        if ans == -1:
            ans = ele
        else:
            ans = min(ans,ele)
print(ans)
```
