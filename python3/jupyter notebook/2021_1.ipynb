```python
#https://www.acmicpc.net/problem/3085
#사탕 게임

import sys

input = sys.stdin.readline

n = int(input())
a = [list(input().strip()) for i in range(n)]


def cal(a):
    n = len(a)
    ret = 1
    for i in range(n):
        cnt = 1
        for j in range(1,n):
            if a[i][j] == a[i][j-1]:
                cnt += 1
            else:
                cnt = 1
            ret = max(cnt, ret)
        cnt = 1

        for j in range(1,n):
            if a[j][i] == a[j-1][i]:
                cnt += 1
            else:
                cnt = 1
            ret = max(cnt, ret)
    return ret


ans = -1
for i in range(n):
    for j in range(n):
        if i+1<n:
            a[i][j], a[i+1][j] = a[i+1][j], a[i][j]
            ans = max(ans, cal(a))
            a[i + 1][j], a[i][j] = a[i][j], a[i+1][j]
        if j+1<n:
            a[i][j], a[i][j+1] = a[i][j+1], a[i][j]
            ans = max(ans, cal(a))
            a[i][j+1], a[i][j] = a[i][j], a[i][j+1]
print(ans)
```

```python
#https://www.acmicpc.net/problem/1107
#리모컨

import sys
input = sys.stdin.readline

n = int(input())
m = int(input())
x = [False]*10
if m:
    dont = list(map(int, input().split()))
    for c in dont:
        x[c] = True

def check(c):
    if c == 0:
        if x[c]:
            return 0
        return 1
    ret = 0
    while c>0:
        if x[c%10]:
            return 0
        ret += 1
        c //= 10
    return ret

ans = abs(100 - n)

for i in range(1000001):
    ret = check(i)
    if ret != 0:
        ans = min(ans, ret + abs(n - i))
print(ans)
```

```python
#https://www.acmicpc.net/problem/14500
#테트로미노

import sys
input = sys.stdin.readline

n, m = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]
visit = [[False]*m for _ in range(n)]

move = [(1,0),(-1,0),(0,1),(0,-1)]
global ans
ans = -1


def go(x, y, cnt, ret):
    if cnt == 3:
        global ans
        ans = max(ans, ret)
        return
    for k in move:
        nx = x + k[0]
        ny = y + k[1]
        if nx<0 or ny<0 or nx>=n or ny>=m:
            continue
        if not visit[nx][ny]:
            visit[nx][ny] = True
            go(nx, ny, cnt +1, ret + a[nx][ny])
            visit[nx][ny] = False


# + 모양에서 한 귀퉁이 떼내기
def go1(x,y):
    global ans
    ret = a[x][y]
    cnt = 0
    for k in move:
        nx = x + k[0]
        ny = y + k[1]
        if nx < 0 or ny < 0 or nx >= n or ny >= m:
            continue
        ret += a[nx][ny]
        cnt += 1
    if cnt == 3:
        ans = max(ans, ret)
    elif cnt > 3:
        for k in move:
            nx = x + k[0]
            ny = y + k[1]
            ans = max(ans, ret - a[nx][ny])


for i in range(n):
    for j in range(m):
        visit[i][j] = True
        go(i,j,0,a[i][j])
        go1(i,j)
        visit[i][j] = False
        
print(ans)
```
