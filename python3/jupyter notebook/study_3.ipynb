```python
#https://www.acmicpc.net/problem/11444
#피보나치 수 6

import sys
input = sys.stdin.readline

mod = 1000000007


def fibo(n):
    if n == 1:
        return [[1,1],[1,0]]
    else:
        prev = fibo(n//2)
        now = [[0,0],[0,0]]
        now[0][0] = (prev[0][0] * prev[0][0] + prev[0][1] * prev[1][0]) % mod
        now[0][1] = (prev[0][0] * prev[0][1] + prev[0][1] * prev[1][1]) % mod
        now[1][0] = (prev[1][0] * prev[0][0] + prev[1][1] * prev[1][0]) % mod
        now[1][1] = (prev[1][0] * prev[0][1] + prev[1][1] * prev[1][1]) % mod

        if n&1:
            prev[0][0] = (now[0][0] + now[1][0]) % mod
            prev[0][1] = (now[0][1] + now[1][1]) % mod
            prev[1][0] = now[0][0]
            prev[1][1] = now[0][1]
            return prev
        return now

n = int(input())
ans = fibo(n)
print(ans[0][1])
```

```python
#https://www.acmicpc.net/problem/17136
#색종이 붙이기

import sys
input = sys.stdin.readline

cnt = [0]*6
ary = [list(map(int, input().split())) for _ in range(10)]

global ans
ans = 987654321


def check(x,y,n):
    if x + n > 10 or y + n > 10:
        return False

    for i in range(n):
        for j in range(n):
            if ary[x+i][y+j] == 0:
                return False
    return True


def setter(x,y,n,op):
    for i in range(n):
        for j in range(n):
            ary[x+i][y+j] = op


def go(x,y,ret):
    global ans
    if ret >= ans:
        return

    flag = False
    while x<10:
        while y<10:
            if ary[x][y] == 1:
                flag = True
                for k in range(5,0,-1):
                    if cnt[k] < 5 and check(x,y,k):
                        setter(x,y,k,0)
                        cnt[k] += 1
                        go(x,y,ret + 1)
                        setter(x,y,k,1)
                        cnt[k] -= 1
                return
            y += 1
        x += 1; y = 0;

    if not flag:
        ans = ret


go(0,0,0)


if ans == 987654321:
    ans = -1
print(ans)
```

```python
#https://www.acmicpc.net/problem/2261
#가장 가까운 두 점
#include <iostream>
#include <algorithm>
#include <vector>
#include <set>
#include <cmath>
using namespace std;

typedef struct ii
{
	int x, y;
}ii;

int dist(ii u, ii v)
{
	return (u.x - v.x)*(u.x - v.x) + (u.y - v.y)*(u.y - v.y);
}
bool cmp(ii u, ii v)
{
	return u.x < v.x;
}
bool operator<(ii u, ii v)
{
	if(u.y!=v.y)
	return u.y < v.y;

	return u.x < v.x;
}

ii a[100001];
int main() 
{
	ios::sync_with_stdio(0);
	cin.tie(0); cout.tie(0);

	int n; cin >> n;
	for (int i = 0; i < n; i++)
		cin >> a[i].x >> a[i].y;

	sort(a, a + n, cmp);
	set<ii>candi = { a[0],a[1], };

	int ans = dist(a[0], a[1]);
	int start = 0;
	for (int i = 2; i < n; i++)
	{
		ii now = a[i];
		while (start < i)
		{
			auto p = a[start];
			int x = p.x - now.x;
			if (x*x > ans)
			{
				candi.erase(p);
				start++;
			}
			else
				break;
		}
		int d = (int)sqrt((double)ans) + 1;
		auto lp = ii({ -10001, now.y - d });
		auto hp = ii({ 10001, now.y + d });

		auto lower = candi.lower_bound(lp);
		auto upper = candi.upper_bound(hp);

		for (auto it = lower; it != upper; ++it)
		{
			auto p = *it;
			ans = min(ans, dist(now, p));
		}
		candi.insert(now);
		
	}
	cout << ans << '\n';
}
```

```python
#시멘트 양 계산하기

#idea : 이전 블록보다 크냐 작냐, 이전 피크값 보다 크냐 작냐
#걍 스택

import sys
input = sys.stdin.readline

day = int(input())
width = int(input())
block = [list(map(int, input().split())) for _ in range(day)]

def solution(day, width, block):
    ans = 0
    a = [0]*width
    for today in block:
        a[0] += today[0]
        st = []

        st.append(0)
        for i in range(1,width):
            a[i] += today[i]
            if a[st[-1]] > a[i]:
                for j in range(st[-1]+1,i):
                    if a[j]<a[i]:
                        ans += a[i] - a[j]
                        a[j] = a[i]
            else:
                val = st.pop()
                for j in range(val+1,i):
                    if a[val] > a[j]:
                        ans += a[val] - a[j]
                        a[j] = a[val]
                st.append(i)
    return ans

print(solution(day, width, block))



```
